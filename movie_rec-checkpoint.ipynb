```python
import pandas as pd

#reading our data : 
movies = pd.read_csv("movies.csv")
```

```python
movies
```

```python
import re
# these are regular expression ( very powerful string matching )

# removing all unneccessary characters like () - _ $
def clean_title(title):
    return re.sub("[^a-zA-Z0-9 ]", "", title)
```

```python
# making a new col in our data frame : 
movies["clean_title"] = movies["title"].apply(clean_title)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a string of all genres concatenated together
all_genres = ' '.join(movies['genres'])

# Generate the word cloud with bright colored text
wordcloud = WordCloud(width=800, height=400, background_color='black', colormap='tab10').generate(all_genres)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# so now we have our clean title 

# now using sklearn -- some inbuilt math functions 

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range = (1,2))

tfidf = vectorizer.fit_transform(movies["clean_title"])

```

```python
# now creating a search function : 
# for that we have to cal similarity between any two names 

# for this we use cosine similarity 
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search(title):

    # title = "Toy Story 1995"
    title = clean_title(title) # cleaning is neccessary
        # user can also enter something ugly
    
    query_vec = vectorizer.transform([title])
    similarity = cosine_similarity(query_vec , tfidf).flatten()
    
    #  now we need to find pairs with highest similarity : 
    indices = np.argpartition(similarity , -5)[-5:]
    
    # now fetching the result based on indices : 
    results = movies.iloc[indices][::-1]

    return results

# that was our searching function
```

```python
# query_vec # so this is a sparse matrix that is what we needed

# print(indices)
# print(results)

import ipywidgets as widgets
from IPython.display import display
#  this are ui libs of jupyter notebook itself for input and ouput ;


# this is our input widget
movie_input = widgets.Text(
    value = "",
    description = "Movie Title : ",
    disabled = False
)

# now creating output widget : 
movie_list = widgets.Output()

def on_type(data):
    with movie_list:
        movie_list.clear_output()  # clear all past ouputs 
        title = data["new"]
        if(len(title) > 5):
            display(search(title))


movie_input.observe(on_type , names = 'value')

display(movie_input, movie_list)
```

```python
def find_similar_movies(movie_id):
    # first we need to find users similar to us 
    similar_users = ratings[(ratings["movieId"] == movie_id) & (ratings["rating"] > 4)]["userId"].unique()
    similar_user_recs = ratings[(ratings["userId"].isin(similar_users)) & (ratings["rating"] > 4)]["movieId"]

    # fetching out the top 10% of recomm.
    similar_user_recs = similar_user_recs.value_counts() / len(similar_users)
    similar_user_recs = similar_user_recs[similar_user_recs > .10]

    # now we need to find how common were these  recommendations : 
    all_users = ratings[(ratings["movieId"].isin(similar_user_recs.index)) & (ratings["rating"] > 4)]
    all_users_recs = all_users["movieId"].value_counts() / len(all_users["userId"].unique())

    
    rec_percentages = pd.concat([similar_user_recs, all_users_recs] , axis=1)
    rec_percentages.columns = ["similar", "all"]
    
    rec_percentages["score"] = rec_percentages["similar"] / rec_percentages["all"]

    rec_percentages = rec_percentages.sort_values("score", ascending = False)
    
    
    return rec_percentages.head(10).merge(movies, left_index = True, right_on = "movieId")[["score","title","genres"]]
```

```python

#  finally we have to make an interactive widget that's it
movie_name_input = widgets.Text(
    value = "",
    description = "Movie Title : ",
    disabled = False
)

recommendation_list = widgets.Output()

def on_type(data):
    with recommendation_list:
        recommendation_list.clear_output()
        title = data["new"]
        if(len(title) > 3):
            results = search(title)
            movie_id = results.iloc[0]["movieId"]
            display(find_similar_movies(movie_id))

movie_name_input.observe(on_type, names = "value")
display(movie_name_input, recommendation_list)
```
